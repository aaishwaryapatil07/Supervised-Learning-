## The Boston Housing Dataset

This dataset contains information collected by the U.S Census Service concerning housing in the area of Boston Mass. 
The dataset is small in size with only 506 cases.<br>

The data was originally published by Harrison, D. and Rubinfeld, D.L. Hedonic prices and the demand for clean air',J. Environ. Economics & Management, vol.5, 81-102, 1978. <br>

**Miscellaneous Details** <br>
-Origin: The origin of the boston housing data is Natural.<br>
-Usage: This dataset may be used for Assessment.<br>
-Number of Cases: The dataset contains a total of 506 cases.<br>
-Order:  The order of the cases is mysterious.<br>
-Variables: There are 14 attributes in each case of the dataset. They are:<br>
-CRIM: per capita crime rate by town<br>
-ZN: proportion of residential land zoned for lots over 25,000 sq.ft.<br>
-INDUS: proportion of non-retail business acres per town.<br>
-CHAS:Charles River dummy variable (1 if tract bounds river; 0 otherwise)<br>
-NOX: nitric oxides concentration (parts per 10 million)<br>
-RM:  average number of rooms per dwelling<br>
-AGE: proportion of owner-occupied units built prior to 1940<br>
-DIS: weighted distances to five Boston employment centres</br>
-RAD: index of accessibility to radial highways<br>
-TAX: full-value property-tax rate per dollar 10,000 <br>
-PTRATIO: pupil-teacher ratio by town<br>
-B: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town<br>
-LSTAT:  percentage lower status of the population <br>
-MEDV: Median value of owner-occupied homes in $1000's<br>
    


### Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_boston
```

```python
boston =load_boston()
```

### Creating Dataframes from the given data

```python
df = pd.DataFrame(data=boston['data'], columns=boston['feature_names'])
```

### Standard Scaling of data

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
scaled_array = ss.fit_transform(df.drop(['Target','RAD','CHAS','AGE'], axis=1))
```

```python
y=df['Target']
```

```python
##looks like RAD, CHAS, AGE to be removed?????
sns.heatmap(df.corr(), cmap='coolwarm')
```

### Creating a PCA of the for proper visualisation

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(scaled_array)
```

```python
x_pca.shape
```

```python
X = pd.DataFrame(data=x_pca, columns = ['A','B'])
```

```python
X.head()
```

```python
plt.scatter(x_pca[:,0], x_pca[:,1], c=y)
```

## Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)
```

```python
X_train.info()
```

```python
X_test.info()
```

## Training the model by different algorithm

#### 1. Logistics Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
ypred = lr.predict(X_test)
```
