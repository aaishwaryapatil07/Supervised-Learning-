## Wine Data Dataset

The wine dataset contains the results of a chemical analysis of wines grown in the same region in Italy but derived from three different cultivars. The analysis determined the quantities of 13 constituents found in each of the three types of wines.

Data Set Characteristics:<br>
    :Number of Instances: 178 (50 in each of three classes)<br>
    :Number of Attributes: 13 numeric, predictive attributes and the class<br>
    :Attribute Information:<br>
 		- 1) Alcohol
 		- 2) Malic acid
 		- 3) Ash
		- 4) Alcalinity of ash  
 		- 5) Magnesium
		- 6) Total phenols
 		- 7) Flavanoids
 		- 8) Nonflavanoid phenols
 		- 9) Proanthocyanins
		- 10)Color intensity
 		- 11)Hue
 		- 12)OD280/OD315 of diluted wines
 		- 13)Proline
        
       	- class:
               - class_0
               - class_1
               - class_2
                
   :Missing Attribute Values: None <br>
    :Class Distribution: class_0 (59), class_1 (71), class_2 (48)<br>
    :Creator: R.A. Fisher<br>
    :Donor: Michael Marshall (MARSHALL%PLU@io.arc.nasa.gov)<br>
    :Date: July, 1988  <br>

### Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importing Wine Dataset from sklearn

```python
from sklearn.datasets import load_wine
```

```python
ml = load_wine()
```

```python
ml.keys()
```

### Creating Dataframes from the given data

```python
df = pd.DataFrame(data= ml['data'], columns = ml['feature_names'])
```

```python
df['Target']=ml['target']
```

```python
df.head()
```

#### Check the head of dataframe, and check out its info() and describe() methods.

```python
df.info()
```

```python
df.describe()
```

### Exploratory Data Analysis

```python
sns.pairplot(df)
```

### Standard Scaling of data

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
scaled_array = ss.fit_transform(df.drop('Target', axis=1))
```

```python
X = pd.DataFrame(data=scaled_array, columns = df.columns[:-1])
```

```python
y = df['Target']
```

```python
sns.heatmap(df.corr(), cmap='coolwarm')
```

### Creating a PCA of the for proper visualisation

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(scaled_array)
```

```python
plt.scatter(x_pca[:,0], x_pca[:,1], c=y)
```

## Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

## Training the model by different algorithm

### 1. Logistics Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
ypred = lr.predict(X_test)
```

```python
ypred
```

#### Checking accuracy using confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, ypred)
```

#### Checking mean absolute error

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y_test, ypred)
```

This shows with Logistics Regression algorithm we could get very less mean error

### 2. Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(X_test, y_test)
```

```python
ypred_svc = svc.predict(X_test)
```

#### Checking accuracy using confusion matrix

```python
confusion_matrix(y_test,ypred_svc)
```

#### Checking mean absoulte error

```python
metrics.mean_absolute_error(y_test,ypred_svc)
```

The above confusion matrix and mean absolute error calculation shows, SVC is the best algorithm for the Wine dataset

### 3. K-nearest Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error =[]
for i in range(1,21):
    kn = KNeighborsClassifier(n_neighbors=i)
    kn.fit(X_train,y_train)
    ypred = kn.predict(X_test)
    error.append(np.mean(ypred!=y_test))
```

```python
plt.plot(range(1,21), error)
```

```python
kn = KNeighborsClassifier(n_neighbors=2)
kn.fit(X_train,y_train)
ypred_kn = kn.predict(X_test)
```

#### Checking accuracy using confusion matrix

```python
confusion_matrix(y_test, ypred_kn)
```

#### Checking mean absoulte error

```python
metrics.mean_absolute_error(y_test, ypred_kn)
```

Above values of mean absolute error shows the prediction by K-nerest is also good
