
Diabetes dataset
================

Notes
-----

Ten baseline variables, age, sex, body mass index, average blood
pressure, and six blood serum measurements were obtained for each of n =
442 diabetes patients, as well as the response of interest, a
quantitative measure of disease progression one year after baseline.

Data Set Characteristics:

  **Number of Instances:** 442

  **Number of Attributes:** First 10 columns are numeric predictive values

  **Target:** Column 11 is a quantitative measure of disease progression one year after baseline (**distributional problem**)

  **Attributes:**<br>
    :Age:<br>
    :Sex:<br>
    :Body mass index:<br>
    :Average blood pressure:<br>
    :S1:<br>
    :S2:<br>
    :S3:<br>
    :S4:<br>
    :S5:<br>
    :S6:<br>
    
Note: Each of these 10 feature variables have been mean centered and scaled by the standard deviation times `n_samples` (i.e. the sum of squares of each column totals 1).


## Imports
**Import pandas, numpy, matplotlib,and seaborn.Then set %matplotlib inline

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Get the Data

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes.keys()
```

```python
df = pd.DataFrame(diabetes['data'], columns= diabetes['feature_names'])
```

```python
df['target'] = diabetes['target']
```

**Check the head of diab, and check out its info() and describe() methods.**

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

# Exploring Datasets

```python
sns.pairplot(df)
```

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
scaled_array = ss.fit_transform(df.drop('target',axis = 1))
```

```python
data = pd.DataFrame(data=scaled_array , columns=df.columns[:-1])
```

```python
y = df['target']
```

```python
data.head()
```

# Data Visualization

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(scaled_array)
```

```python
x_pca.shape
```

```python
plt.scatter(x_pca[:,0] , x_pca[:,1], c=y)
```

**Note:** Data is very scattered

# Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,df['target'],test_size=0.3)
```

# 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
ypred_lr = lr.predict(X_test)
```

**Plotting graph**  

```python
plt.scatter(ypred_lr,y_test,edgecolors='b')
```

**Predicting mean error**

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test,ypred_lr)
```

# 2. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
cl = DecisionTreeRegressor()
```

```python
cl.fit(X_train,y_train)
```

```python
ypred_dt = cl.predict(X_test)
```

**Plotting graph** 

```python
plt.scatter(ypred_dt,y_test,edgecolors='b')
```

**Predicting mean error**

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test,ypred_dt)
```

# 3. Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
ypred_rf = rf.predict(X_test)
```

**Plotting Graph**

```python
plt.scatter(y_test, ypred_rf, edgecolors='r')
```

**Predicting mean error**

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test,ypred_rf)
```

# Conclusion: 

**mean_absolute_error** predicted by different model:<br>

1. Linear Regression: 48.70
